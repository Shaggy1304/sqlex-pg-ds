```python
import pandas as pd
import psycopg2

con = psycopg2.connect(host='this_postgres',
                       user='postgres',
                       password='password',
                       database='postgres')

con.autocommit = True

def select(sql):
    return pd.read_sql(sql,con)
```

List all tables

```python
sql = '''SELECT * FROM pg_catalog.pg_tables t
where t.schemaname = 'public'
'''
```

```python
select(sql)
```

Get hard drive capacities that are identical for two or more PCs.
Result set: hd.

```python
sql = '''
select t.hd from pc t
group by t.hd
having count(1) > 1
'''
```

```python
select(sql)
```
